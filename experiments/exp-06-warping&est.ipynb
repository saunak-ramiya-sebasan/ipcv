# Exp-06 Image Warping n Estimatinon

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('fox.png')

if img is None:
    print("Error loading image")
    exit()

# ---------------------------
# Define 4 points in original image
# (manually choose points)
# ---------------------------
pts1 = np.float32([
    [100, 100],   # top-left
    [400, 100],   # top-right
    [100, 400],   # bottom-left
    [400, 400]    # bottom-right
])

# ---------------------------
# Define where those points go (output)
# ---------------------------
pts2 = np.float32([
    [0, 0],
    [300, 0],
    [0, 300],
    [300, 300]
])

# ---------------------------
# Compute transformation matrix
# ---------------------------
matrix = cv2.getPerspectiveTransform(pts1, pts2)

# ---------------------------
# Apply warping
# ---------------------------
warped = cv2.warpPerspective(img, matrix, (300, 300))

# Convert BGR → RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
warped_rgb = cv2.cvtColor(warped, cv2.COLOR_BGR2RGB)

# Show results
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')
plt.show()

plt.imshow(warped_rgb)
plt.title("Warped Image")
plt.axis('off')
plt.show()
```
