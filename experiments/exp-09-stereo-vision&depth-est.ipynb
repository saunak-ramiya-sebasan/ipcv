# Exp-09 Stereo Vision N Depth Estimation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load left and right images (grayscale)
imgL = cv2.imread('left.png', 0)
imgR = cv2.imread('right.png', 0)

if imgL is None or imgR is None:
    print("Error loading images")
    exit()

# Create Stereo Matcher
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)

# Compute disparity (depth map)
disparity = stereo.compute(imgL, imgR)

# Normalize for display
disp_norm = cv2.normalize(disparity, None, 0, 255, cv2.NORM_MINMAX)
disp_norm = np.uint8(disp_norm)

# Show results (matplotlib)
plt.imshow(imgL, cmap='gray')
plt.title("Left Image")
plt.axis('off')
plt.show()

plt.imshow(imgR, cmap='gray')
plt.title("Right Image")
plt.axis('off')
plt.show()

plt.imshow(disp_norm, cmap='gray')
plt.title("Depth Map (Disparity)")
plt.axis('off')
plt.show()
```
